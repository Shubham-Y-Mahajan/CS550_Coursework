# Loading the data

```python
import pandas as pd
```

```python
df=pd.read_csv('WineQT.csv')
```

```python
df
```

# Applying Transformation on the 'quality' Column

```python
df['quality'] = df['quality'].apply(lambda x: 0 if x <= 6 else 1)
```

```python
df
```

# Normalizing all other attributes using Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df.iloc[:, :-2] = scaler.fit_transform(df.iloc[:, :-2])  # Normalize all columns except 'quality' and 'Id'
```

```python
df
```

# Saving the Data Frame to a new .csv file

```python
df.to_csv('wine_quality_normalized.csv', index=False)
```
